```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from graph_tool.all import *
from sklearn.preprocessing import MinMaxScaler
```

## 資料讀取

```python
df = pd.read_excel("topo.xlsx")
col_fil = list(filter(lambda x: ('Unnamed' in x) | (x=='1.你這次接觸者的姓名？（若不知道，則填NO)'), df.columns))
df = df.drop(col_fil, axis = 1).rename(columns={'使用者': 'user_main', '被訪談者': 'user_sub', '目的':'purpose', 
                                                '時間戳':'timestamp'})
df.head()
```

```python
mapping = pd.read_csv('mapping.csv')
mapping = mapping.T.reset_index().T.reset_index().drop('index', axis = 1).rename(columns={0:'map', 1:'name'})
mapping.head()
```

```python
# 處理日期轉換星期
df['weekday'] = df['timestamp'].apply(lambda x: pd.Timestamp(x).weekday())
df = df.drop('timestamp', axis = 1)
```

## 資料清理

```python
def final_clean(df_specific):
    
    arr = []
    def user_clean(r):
        if r['who_first'] != 1:
            r['user_clean'] = r['user_sub']
            r['user_sub'] = r['user_main']
            r['user_main'] = r['user_clean']

            if r['who_first'] in [3, 4]:
                list_test = df_specific.iloc[r['row_index']].to_list()
                arr.append(list_test)
        return r
    
    # map name
    df_specific = df_specific.merge(mapping, how='left', left_on='user_main', right_on='map')
    df_specific['user_main'] = df_specific['name']
    
#     df_specific['user_num'] = df_specific['map'].apply(lambda x: int(x[1:]))
    df_specific = df_specific.drop(['map', 'name'], axis = 1)
    
    # 處理 使用者交互狀態
    df_specific['user_clean'] = df_specific['user_main']
    df_specific['row_index'] = df_specific.index
    df_specific = df_specific.apply(lambda r: user_clean(r), axis = 1)
    df_temp = pd.DataFrame(arr, columns = df_specific.columns)
    df_specific = pd.concat([df_temp,df_specific],ignore_index=True)

    
    # 新增 權重欄位
    df_specific['familiar'] = df_specific['familiar'].apply(
        lambda x: 0.25 if x == 4 else (0.5 if x == 3 else (0.75 if x == 2 else 1)))
    df_specific['importance'] = df_specific['importance'].apply(
        lambda x: 0.25 if x == 4 else (0.5 if x == 3 else (0.75 if x == 2 else 1)))
    df_specific['like_or_not'] = df_specific['like_or_not'].apply(
        lambda x: 0.25 if x == 4 else (0.5 if x == 3 else (0.75 if x == 2 else 1)))
    df_specific['weight'] = df_specific['familiar'] + df_specific['importance'] + df_specific['like_or_not']

    # 刪除 不需要的欄位
    df_specific = df_specific.drop(['familiar', 'importance', 'like_or_not', 'who_first', 'user_clean', 'row_index'], axis = 1)
    df_specific = df_specific.dropna()

    # # 權重 = 權重＊次數
    # # df['weight'] = df['weight'] * df['count']
    # # df = df.drop('count', axis = 1)

    # # 處理編碼
    df_specific['user_sub'] = df_specific['user_sub'].apply(lambda x: x.encode('utf-8').decode('utf-8'))
    df_specific['user_main'] = df_specific['user_main'].apply(lambda x: x.encode('utf-8').decode('utf-8'))
    
    return df_specific
```

## 整理資料放入 graph

```python
def gen_data(df_specific, types=0, df_all = df):
    
    # 處理數量
#     df_specific['count'] = df_specific.groupby(['user_main', 'user_sub', 'weight', 'purpose', 'user_num']).transform('size')
    df_specific = df_specific.drop(['purpose', 'weekday'], axis = 1)
    df_specific['count'] = df_specific.groupby(['user_main', 'user_sub', 'weight']).transform('size')
    df_specific = df_specific.drop_duplicates()

    # 所有 user 名稱
    # 所有報告的 node 排序依據 AX 編號 
#     取出所有名稱 -> 轉換df -> 對應 AX 編號 -> 取最大數並重新編號 -> 依據編號排序
    uni_user_uni = pd.concat([df_specific['user_main'], df_specific['user_sub']]).unique()
    uni_user_df = pd.DataFrame(uni_user_uni, columns = ['name'])
    uni_user_df = uni_user_df.merge(mapping, how='left', left_on='name', right_on='name')
    max_idx = max(mapping['map'].apply(lambda x: int(x[1:])))
    uni_user_df['num'] = uni_user_df['map'].apply(lambda x: max_idx+1 if x!=x else int(x[1:]))
    uni_user_df = uni_user_df.sort_values('num')
    uni_user = uni_user_df.reset_index().drop('index', axis = 1).reset_index()['name']
    
    
#     所有編號跟全圖一致
    if types!=0:
        uni_user_all = pd.concat([df_all['user_main'], df_all['user_sub']]).unique()
        uni_user_all_df = pd.DataFrame(uni_user_all, columns = ['name'])
        uni_user_all_df = uni_user_all_df.merge(mapping, how='left', left_on='name', right_on='name')
        uni_user_all_df['num'] = uni_user_all_df['map'].apply(lambda x: max_idx+1 if x!=x else int(x[1:]))
        uni_user_all_df = uni_user_all_df.sort_values('num')
        
        map_index_name = uni_user_all_df.reset_index().drop('index', axis = 1).reset_index()[['index', 'name']]
        
        merge_index = uni_user_df.merge(map_index_name, how='left', left_on='name', right_on='name')
        
        uni_index = merge_index['index']
        
    else:
        uni_index = uni_user_df.reset_index().drop('index', axis = 1).reset_index()['index']
        
    

    # 所有 data 轉換陣列
    data_list = np.array(df_specific[['user_main', 'user_sub', 'weight', 'count']])

    # 所有 user 列表
    user_list = data_list[:, :2]

    # 所有 權重 列表
#     weight_list = data_list[:, 2:]

    # edge 列表轉換格式
    edge_idx = [[np.where(uni_user==y)[0][0] for y in x] for x in user_list]
    edge_list = list(map(tuple, edge_idx))
    
    return uni_user, uni_index, edge_list, df_specific
```

## 建立 graph

```python
def gen_graph(uni_user, edge_list):
    # graph
    g = Graph()

    g.add_vertex(len(uni_user))
    g.add_edge_list(edge_list)

    deg = g.degree_property_map("total")
    
    control = g.new_edge_property("vector<double>")

    in_arr = g.degree_property_map("in").get_array()
    in_arr_reshape = in_arr.reshape(in_arr.shape[0], 1)
    
    return deg, control, in_arr_reshape, g
```

```python
# idx 名稱 同顏色 分大小¶
def gen_graph_diff_size(deg, control, in_arr_reshape, uni_index, g, output_size, file_name):
    # scale 入度 for node size
    scaler_vertex_size = MinMaxScaler(feature_range=(16, 48))
    scale_fit_vertex_size = scaler_vertex_size.fit_transform(in_arr_reshape)
    
    # node index
    v_prop_index = g.new_vertex_property('double')
    g.vertex_properties["vertex_index"] = v_prop_index
    for idx, val in enumerate(uni_index):
        g.vp.vertex_index[g.vertex(idx)] = val

    # node size
    v_prop_size = g.new_vertex_property('double')
    g.vertex_properties["vertex_size"] = v_prop_size
    for idx, val in enumerate(scale_fit_vertex_size):
        g.vp.vertex_size[g.vertex(idx)] = val[0]

    # scale 入度 for node font size
    scaler_vertex_font_size = MinMaxScaler(feature_range=(4, 8))
    scale_fit_vertex_font_size = scaler_vertex_font_size.fit_transform(in_arr_reshape)

    # node font size
    v_prop_font_size = g.new_vertex_property('double')
    g.vertex_properties["vertex_font_size"] = v_prop_font_size
    for idx, val in enumerate(scale_fit_vertex_font_size):
        g.vp.vertex_font_size[g.vertex(idx)] = val[0]
        
    # node vertex_text
    v_prop_vertex_text = g.new_vertex_property('string')
    g.vertex_properties["vertex_text"] = v_prop_vertex_text
    for idx, val in enumerate(uni_index):
        g.vp.vertex_text[g.vertex(idx)] = str(val)
        
    
    # draw
    graph_draw(g, vertex_size=g.vp.vertex_size, 
               vertex_font_size=g.vp.vertex_font_size,
               vertex_color='grey', 
               vertex_fill_color='skyblue',
#                vertex_text = g.vp.vertex_index,
               vertex_text = g.vp.vertex_text,
               vertex_text_color = 'black',
               vertex_font_family='Heiti TC',
               edge_pen_width=1,
               edge_control_points=control,
               vorder=deg,
               bg_color = 'white',
               output_size=(output_size, output_size), output=file_name)
```

```python
# idx 名稱 同大小 分顏色
def gen_graph_diff_color(deg, control, uni_index, g, output_size, file_name):
    control = g.new_edge_property("vector<double>")

    vertex_text_color = []
    for i in prop_to_size(deg, 0, 100, power=.1).get_array():
        if i > 50:
            vertex_text_color.append('1')
        else:
            vertex_text_color.append('0')
            
    prop_to_size(deg, 100, 0, power=.9)
    
    # node vertex_text
    v_prop_vertex_text = g.new_vertex_property('string')
    g.vertex_properties["vertex_text"] = v_prop_vertex_text
    for idx, val in enumerate(uni_index):
        g.vp.vertex_text[g.vertex(idx)] = str(val)
    
    graph_draw(g, vertex_font_size=8,
               vertex_text = g.vp.vertex_text,
#                vertex_text_color = prop_to_size(deg, 100, 0, power=.9),
#                vertex_text_color = "auto",
               vertex_font_family='Heiti TC',
               vertex_fill_color=prop_to_size(deg, 0, 100, power=.1),
               vertex_size=prop_to_size(deg, 22, 24, power=.2), 
               vcmap=plt.cm.Oranges,
               edge_pen_width=1,
               edge_control_points=control,
               vorder=deg,
               bg_color = 'white',
               output_size=(output_size, output_size), output=file_name)
```

## 建立報告

```python
def gen_report(g, df_specific, uni_user, uni_index, file_name_node, file_name_edge):
    # 產出 node 報告
    # 出入度
    in_list = np.array(g.degree_property_map("in").get_array())
    out_list = np.array(g.degree_property_map("out").get_array())

    # 介數中心度
    vp, ep = betweenness(g)
    betweenness_list = np.array(vp.get_array())

    # 接近中心度
    closeness_list = np.array(closeness(g).get_array())
    closeness_list = np.nan_to_num(closeness_list)

    # 平均路徑長度
    dist = shortest_distance(g)
    sd = []
    for d_arr in dist:
        for i in d_arr:
            if i==2147483647: sd.append(0)
            else: sd.append(i)
    ave_path_length = sum(sd)/(g.num_vertices()**2-g.num_vertices())
    
    # report df
    df_report = pd.DataFrame({'對應數字':uni_index,
                              '使用者':uni_user, 
                              '入度':np.round(in_list, 4), 
                              '出度':np.round(out_list, 4),
                              '介數中心度':np.round(betweenness_list, 4),
                              '接近中心度':np.round(closeness_list, 4), 
                              '全域集聚系数':np.round(global_clustering(g)[0], 4), 
                              '地方集聚系数':np.round(np.array(local_clustering(g).get_array()), 4), 
                              '平均路徑長度':np.round(ave_path_length, 4)})
#     df_report = df_report.reset_index(names='對應數字')

    # 度中心度
    arr_in_out = np.array(df_report[['入度', '出度']])
    degree_central = [i.sum()/arr_in_out.sum() for i in arr_in_out]
    df_report['度中心度'] = np.round(degree_central, 4)

    # report csv
    df_report.to_csv(file_name_node)
    
    
    # 產出 edge 報告
#     df_specific = df_specific.drop('purpose', axis = 1)
    
    df_specific.columns = ['主要', '次要', '權重', '數量']
    
    # report csv
    df_specific.to_csv(file_name_edge)
    
    return in_list, out_list, betweenness_list
```

## 產出分佈圖

```python
def gen_hist(uni_user, in_list, out_list, betweenness_list, file_name_deg, file_name_between):
    
    # 度分佈圖（數量）
    
    df_in_out = pd.DataFrame({'uni_user':uni_user, 'in_list':in_list, 'out_list':out_list}).reset_index()

    arr_temp = []
    start = 0
    for i in df_in_out['index']:
        arr_temp.append(start)
        if i in range(0, len(df_in_out['index']), 20)[1:]:
            start+=20

    df_in_out['binss']=arr_temp
    df_in_out['in_out'] = df_in_out['in_list']+df_in_out['out_list']

    in_out = df_in_out.groupby('binss').sum('in_out')['in_out']/20
    in_avg = df_in_out.groupby('binss').sum('in_list')['in_list']/20
    out_avg = df_in_out.groupby('binss').sum('out_list')['out_list']/20

    df_in_out_avg = pd.DataFrame({'in_out':in_out, 'in_list':in_avg, 'out_list':out_avg}).reset_index()

    plt.figure(figsize = (12, 4))
    sns.set_style('whitegrid')


    plt.bar(df_in_out_avg['binss'], df_in_out_avg['in_out'], width=20)
    plt.plot(df_in_out_avg['binss'], df_in_out_avg['in_out'], color='orange')

    plt.xlabel('index')
    plt.ylabel('degree')

    plt.savefig(file_name_deg)
    
    # 介數分佈圖
    
    df_betweenness = pd.DataFrame({'uni_user':uni_user, 'betweenness_list': betweenness_list})
    df_betweenness = df_betweenness.drop(df_betweenness[df_betweenness['betweenness_list']==0].index).reset_index()

    plt.figure(figsize=(32, 4))

    plt.bar(df_betweenness.index, df_betweenness['betweenness_list'])
    plt.plot(df_betweenness.index, df_betweenness['betweenness_list'], color='orange')

    plt.xlabel('index')
    plt.ylabel('betweenness')

    plt.savefig(file_name_between)
```

## 全部 controller

```python

# 整理放入 graph 資料
uni_user, uni_index, edge_list, df_specific = gen_data(df)
print('gen_data done!')

# 建立 graph
deg, control, in_arr_reshape, g = gen_graph(uni_user, edge_list)
print('gen_graph done!')

# idx 名稱 同顏色 分大小
gen_graph_diff_size(deg, control, in_arr_reshape, uni_index, g, 3000, f'topo_idx_size.png')
print('gen_graph_diff_size done!')

# idx 名稱 同大小 分顏色
gen_graph_diff_color(deg, control, uni_index, g, 3500, f'topo_idx_color.png')
print('gen_graph_diff_color done!')

# 產出 node 報告
in_list, out_list, betweenness_list = gen_report(g, df_specific, uni_user, uni_index, 
                                                 f'topo_report.csv', 
                                                 f'topo_between_report.csv')
print('gen_report done!')

# 分佈圖
gen_hist(uni_user, in_list, out_list, betweenness_list, 
         f'degree_distribution.png', 
         f'betweenness_distribution.png')
print('gen_hist done!')
```
